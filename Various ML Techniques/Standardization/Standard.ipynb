```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
age=df.iloc[:,2]
```

```python
salary=df.iloc[:,3]
```

```python
salary
```

```python
age
```

```python
# First calulating mean for age
sum=0
for i in range(0,400):
    sum+=age[i]

mean_age=sum/400
print("Mean of age is ", mean_age)
```

```python
# Calculing the standard deviation
for i in range(0,400):
    sum+=(age[i]-mean_age)**2

var=sum/400
std_age=var**.5
print("Standard deviation for age is ",std_age)
```

```python
## Now trasnforming all the values of age by the formula of standarization

for i in range(0,400):
    x= (age[i]-mean_age)/std_age
    age[i]=x
    print(x)
```

### As we can see that all the values of age has been tranformed, now we will perform the same operation on the salary column
