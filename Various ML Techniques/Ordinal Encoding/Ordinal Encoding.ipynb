```python
# importing essential libraries
import numpy as np
import pandas as pd
```

```python
# importing the dataset
df = pd.read_csv('customer.csv')
```

```python
df.head(10)
```

### As we can see that the column gender, review, education, purchased is in categorial form, our task to transfrom these into numerical data.
### By seeing the data and after understanding it, we came to know that we need to perform ordinal encoding on the "Review " and "Education" column

```python
# As right now we are learning ordinal endcoding so we will only be focusing 
# only on columns such as review,education and purchased.
## Note: On Output column "Purchased" we will be doing "Label Encoding"

df = df.iloc[:,2:]
```

```python
df.head()
```

```python
# Splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=.25)
```

```python
X_train
```

```python
# importing the ordinal encoder 
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Creating an object for Ordinal encoder

## Creating a list of the order of data as we need 
review=['Poor','Average','Good']
education=['School','UG','PG']

## Here poor will be assigned to 0, average as 1 and good as 2
## similarly school as 0, UG as 1 and PG as 2

oe = OrdinalEncoder(categories=[review,education])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train
```

```python
# Now we will use label encoder on output value
# Note: As mentioned in the documentation of Sklearn that 
# This transformer should be used to encode target values, i.e. y, and not the input X, hence we will use label encoder only
# link- https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html
```

```python
# importing the library
from sklearn.preprocessing import LabelEncoder
```

```python
# Creating an object
le = LabelEncoder()

```

```python
le.fit(y_train)
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
