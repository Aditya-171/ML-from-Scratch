```python
# Finding out the min and max value of age
print(age.min())
print(age.max())
```

### Performing MinMax Scaling

```python
# performin minmax Scaling on age
for i in range(0,400):
    x= (age[i]-18)/(60-18)
    age[i]=x
    print(x)
```

```python
# Finding out the min and max value of salary
min_S=salary.min()
max_S=salary.max()
print(min_S)
print(max_S)
```

```python
# performin minmax Scaling on Salary
for i in range(0,400):
    x= (salary[i]-min_S)/(max_S-min_S)
    age[i]=x
    print(x)
```
