### There are various ways to perform this task


## 1. Using pandas

```python
## get_dummies is used in pandas,
## The columns name which need to be transformed should be input as a list into the function
pd.get_dummies(df,columns=['fuel','owner'])
```

```python
## As we can see new columns has been created 
```

```python
## To remove the problem of multi collinearity we will remove the first column
pd.get_dummies(df,columns=['fuel','owner'], drop_first=True)
```
